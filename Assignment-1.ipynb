#### Q4) Write a Python program to find the volume of a sphere with diameter 12 cm.  Formula: V=4/3 * π * r 3 

#### Solution

```python
diameter=12
r=diameter/2
pie=3.14

V=(4/3 * pie * (r**3))

print("volume of a sphere with diameter 12 cm is {}".format(V))
```

#### Q2) Create the below pattern using nested for loop in Python. 
 
 
#### * 
#### * *  
#### * * *
#### * * * *  
#### * * * * *
#### * * * * 
#### * * *  
#### * *  
#### * 

#### Solution-

```python
rows=int(input("Enter number of rows: "))

for i in range(0,rows):
    for j in range(0,i+1):
        print("*",end=" ")
    print()
    
for i in range(rows,0,-1):
    for j in range(0,i-1):
        print("*",end=" ")
    print()
```
